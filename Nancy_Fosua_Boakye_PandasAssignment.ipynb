Q2: Merge all the files and add a new column that defines the region where the country belongs to.

```python
# We adding a new column called region to each datasets
central_data['region'] ='central'
north_data['region'] = 'north'
south_data['region'] = 'south'
east_data['region'] = 'east'
west_data['region'] = 'west'
```

```python
# We rename the columns to have the same column names to easy merging
central_data.rename(columns = { central_data.columns[0]:'Name', central_data.columns[1]:'Area', central_data.columns[2]:'Population', central_data.columns[3]:'Density', central_data.columns[4]:'Capital'}, inplace = True)
north_data.rename(columns = {north_data.columns[0]:'Name',north_data.columns[1]:'Area',north_data.columns[2]:'Population',north_data.columns[3]:'Density', north_data.columns[4]:'Capital'},inplace=True)
south_data.rename(columns = {south_data.columns[0]:'Name',south_data.columns[1]:'Area',south_data.columns[2]:'Population',south_data.columns[3]:'Density',south_data.columns[4]:'Capital'},inplace= True)
east_data.rename(columns={east_data.columns[0]:'Name',east_data.columns[1]:'Area',east_data.columns[2]:'Population',east_data.columns[3]:'Density',east_data.columns[4]:'Capital'},inplace =True)
west_data.rename(columns={west_data.columns[0]:'Name',west_data.columns[1]:'Area',west_data.columns[2]:'Population',west_data.columns[3]:'Density',west_data.columns[4]:'Capital'},inplace = True)
```

```python
Merge_data= pd.concat([central_data, south_data, north_data, west_data, east_data],ignore_index= True)
Merge_data
```

Q3: Calculate and present in the same Dataframe: The mean of the population () of each region, the standard deviation () and the calculation of the coefficient of variation Cv and conclude:
      

```python
m= Merge_data.groupby(['region']).Population.mean()
s_d =Merge_data.groupby(['region']).Population.std()
c_variation = s_d / m
m,s_d ,c_variation
mama = pd.DataFrame({'Population_Mean':m,'Population_stand_Dev':s_d,'Coefficient_variation': c_variation})
mama.reset_index(inplace=True) # to reset the index (region) to column
mama
```

Q5: Obtain the average population of each region and complete with the name of country having the population closest to the average of the corresponding group, on the same line in the Serie or Dataframe obtained.

```python
average = pd.DataFrame( Merge_data.groupby('region').Population.mean())
average
```

```python
def remean_pop(row):
    row = abs(row.Population - average.loc[row.region])
    return row
Merge_data.apply(remean_pop,axis='columns') #
```

```python
s = Merge_data.apply(remean_pop,axis='columns')
t= Merge_data.loc[:, ['Name','region']]
t['Population'] = s
t
```

```python
average = pd.DataFrame( Merge_data.groupby('region').Population.mean())
index_avPop = t.groupby('region').Population.idxmin()
t.loc[index_avPop ,'Name']
average['Name'] = t.loc[index_avPop ,'Name'].values
average.rename(columns={ average.columns[0]:'average_population', average.columns[1]:'closest_country'},inplace=True)
average.reset_index(inplace=True)
average
```
